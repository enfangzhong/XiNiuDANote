## 风格管理
先来看看matplotlib画出来的图啥样子~

```python
%matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
np.random.seed(sum(map(ord, "aesthetics"))) # 给每一个任务单独分配随机种子
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7): # up to but not include
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
sinplot()
```

再来看看seaborn画出来滴
* 色彩柔和
* 灰色背景
* 白色网格

```python
import seaborn as sns
sns.set()
sinplot()
```

* 风格参数设置
* 尺度参数设置

### 风格参数设置
* darkgrid/whitegrid：更关注数值
* dark/white：更关注整体趋势
* ticks：强化刻度

```python
sns.set()
sns.set_style("whitegrid")
sinplot()
```

```python
sns.set_style("ticks")
sinplot()
sns.despine()
```

```python
sinplot()
sns.despine(offset=20)
```

### 尺度参数设置

```python
sns.set_context("paper")
sinplot()
sns.despine()
```

```python
sns.set_context("notebook")
sinplot()
sns.despine()
```

```python
sns.set_context("talk")
sinplot()
sns.despine()
```

```python
sns.set_context("poster")
sinplot()
sns.despine()
```

## 颜色管理
### 分类调色板
* 适用于分类描述的数据，而且分类之间无序
* hls颜色系统，色调+亮度+饱和度（hue + luminance + saturation）
* husl颜色系统，hls中不同颜色存在视差，eg.黄色偏亮，蓝色偏暗，husl使得更加平衡

```python
# 恢复默认设置
sns.set()
current_palette = sns.color_palette()
sns.palplot(current_palette)
sns.set_palette(sns.color_palette("hls", 10))
current_palette = sns.color_palette()
sns.palplot(current_palette)
sns.set_palette(sns.color_palette("husl", 10)) # http://www.hsluv.org
current_palette = sns.color_palette()
sns.palplot(current_palette)
```

### 序列调色板
* 主要通过亮度和饱和度区分有序分类
* 关注单级数据

```python
sns.palplot(sns.color_palette("Greens",10))
```

### 分叉调色板
* 关注两级数据

```python
sns.palplot(sns.color_palette("coolwarm",11))
```
