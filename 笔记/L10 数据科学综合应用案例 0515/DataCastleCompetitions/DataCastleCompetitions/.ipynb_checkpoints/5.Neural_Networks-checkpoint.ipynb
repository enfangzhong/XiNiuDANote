```python
# 必要的库
import tensorflow as tf
import numpy as np
import pandas as pd
#加载数据
X_test = pd.read_pickle('./input/X_test.pkl')
X_train = pd.read_pickle('./input/X_train.pkl')
y_train = pd.read_pickle('./input/y_train.pkl')
```

```python
X_train.columns
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import minmax_scale as mm
X_test = mm(X_test[['成绩排名', '消费金额', '其他', '图书馆', '开水', '教务处', '文印中心', '校医院', '校车', '洗衣房', '淋浴', '超市', '食堂']])
X_train= mm(X_train[['成绩排名', '消费金额', '其他', '图书馆', '开水', '教务处', '文印中心', '校医院', '校车', '洗衣房', '淋浴', '超市', '食堂']])
```

```python
X_train = pd.DataFrame(X_train, columns=['成绩排名', '消费金额', '其他', '图书馆', '开水', '教务处', '文印中心', '校医院', '校车', '洗衣房', '淋浴', '超市', '食堂'])
X_test = pd.DataFrame(X_test, columns=['成绩排名', '消费金额', '其他', '图书馆', '开水', '教务处', '文印中心', '校医院', '校车', '洗衣房', '淋浴', '超市', '食堂'])
```

```python
X_train.head()
```

```python
y_train = pd.DataFrame(y_train)
```

```python
y_train.head()
```

```python
# 将sub 数据换成０，１，２，３
sub_map = {0:0,1000:1,1500:2,2000:3}
def cate_map(c): return sub_map[c]
y_train['助学金金额'] = y_train['助学金金额'].apply(cate_map)
```

```python
TRAINING_DATA = pd.concat([X_train, y_train], axis=1)
```

```python
TRAINING_DATA.head()
```

```python
TRAINING_DATA.columns = ['rank','feature01','feature02','feature03','feature04','feature05','feature06','feature07','feature08','feature09','feature10','feature11','feature12','sub']
```

```python
TRAINING_DATA.head()
```

```python
X_test.columns = ['rank','feature01','feature02','feature03','feature04','feature05','feature06','feature07','feature08','feature09','feature10','feature11','feature12']
```

```python
# 将数据转换成tensor格式方便TF的计算
feature_list = X_test.columns.values.tolist()
def input_fn(df):
    feature_cols  = {k: tf.constant(df[k].values)
                     for k in feature_list}
    label = tf.constant(df['sub'].values)
    return feature_cols, label

def input_fn2(df):
    feature_cols  = {k: tf.constant(df[k].values)
                     for k in feature_list}
    return feature_cols
```

```python
# Specify that all features have real-value data
rank = tf.contrib.layers.real_valued_column("rank")
feature01 = tf.contrib.layers.real_valued_column("feature01")
feature02 = tf.contrib.layers.real_valued_column("feature02")
feature03 = tf.contrib.layers.real_valued_column("feature03")
feature04 = tf.contrib.layers.real_valued_column("feature04")
feature05 = tf.contrib.layers.real_valued_column("feature05")
feature06 = tf.contrib.layers.real_valued_column("feature06")
feature07 = tf.contrib.layers.real_valued_column("feature07")
feature08 = tf.contrib.layers.real_valued_column("feature08")
feature09 = tf.contrib.layers.real_valued_column("feature09")
feature10 = tf.contrib.layers.real_valued_column("feature10")
feature11 = tf.contrib.layers.real_valued_column("feature11")
feature12 = tf.contrib.layers.real_valued_column("feature12")

feature_columns = [rank,feature01,feature02,feature03,feature04,feature05,feature06,feature07,feature08,feature09,feature10,feature11,feature12]
# 创建一个基本的神经网络
classifier = tf.contrib.learn.DNNClassifier(feature_columns=feature_columns,
                                            hidden_units=[12, 12],
                                            n_classes=4,
                                            model_dir="/tmp/sub_model")
```

```python
classifier.fit(input_fn=lambda: input_fn(TRAINING_DATA),steps=20000)
```

```python
y = classifier.predict(input_fn=lambda: input_fn2(X_test))
type(y)
```
