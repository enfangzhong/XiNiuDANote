# 数据可视化

## 可视化的窘境
* 没有给予应有的重视（100%）
* 没有系统全面的讲述（50%）
* 没有足够数量的训练（25%）
---
* 可视化 = 画图表？
* 画完图表 = 大功告成？
* 什么样的图表是好的可视化？
* 什么场景画什么图？

## 可视化的目的：解释数据 + 传递信息
* 比较两组数据的大小
* 校验两组数据的差异
* 统计随机变量的分布
* and so on

## 可视化的原则：Simpler  is almost always better.
高效的表达 = 清晰的思路 = 价值的体现
* 尽量用低维度表达
* 一定避免视觉噪声（标记、注释、颜色）
* 确保各个元素清晰

三类数据：定量数据、定序数据、定类数据
* 定量数据（Quantitative）：连续、离散
* 定序数据（Ordinal）：尺码、态度
* 定类数据（Categorical）：城市、品类

## 可视化 VS 翻译
> 译事三难，信，达，雅。
* 信：准确
* 达：明白
* 雅：审美

### 定量数据

位置 > 长度/角度 > 面积 > 体积 > 密度 > 颜色

### 定序数据

位置 > 密度 > 颜色 > 连接 > 包含

### 定类数据

位置 > 颜色 > 连接 > 包含 > 形状
