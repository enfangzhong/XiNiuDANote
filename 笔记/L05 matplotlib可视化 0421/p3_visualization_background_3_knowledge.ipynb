# 数据分析相关知识梳理

## 数据分析的3个要素

目标 + 方法 + 结果

* 目标：明确性 + 可量化
* 方法：统计分析 + 数据挖掘
* 结果：解释性描述 + 可视化呈现

## 统计分析方法

### 描述性统计分析

通过图表或数学的方法，对数据资料进行整理、分析，并对数据的分布状态、数字特征和随机变量之间的关系进行估计和描述的方法。

#### 集中趋势分析

* 平均数：关注数值，鲁棒性弱
$$
\bar{x} = \frac{\sum_N x}{N}
$$

* 中位数：关注顺序，鲁棒性强
$$
m_{0.5} = x_{(N+1)/2}
$$
$$
m_{0.5} = \frac{x_{N/2}+x_{N/2+1}}{2}
$$

* 众数：关注频次

#### 离中趋势分析

* 全距：范围误差、极差，最大值和最小值的差距，刻画数据离散程度
$$
R = x_{\max} - x_{\min}
$$

* 四分位距：四分差，interquartilerange，刻画“大多数样本”的离散程度，鲁棒性强
$$
IQR = Q_3 - Q_1
$$

--------------------

* 平均差：差的绝对值的均值，1范数，异常值影响
$$
MD = \frac{\sum_N \|x - \bar{x}\|}{N}
$$

* 方差：差的平方的均值，2范数，异常值影响

总体方差
$$
\sigma^2 = \frac{\sum_N (X-E(X))^2}{N}
$$

样本方差
$$
s^2 = \frac{\sum_N (x - \bar{x})^2}{N-1}
$$

* 标准差：方差的算数平方根

> 平均差 VS 方差：对异常值的敏感程度不同

* 离散系数

$$
CV = \frac{\sigma}{\mu}
$$

#### 相关分析

* 正相关 VS 负相关
* 完全相关 VS 不相关
* 线性相关 VS 非线性相关
* 单相关 VS 复相关

$$
r(X, Y) = \frac{cov(X,Y)}{\sqrt{var(X)}\sqrt{var(Y)}}
$$
